# Neo4j Queries 

```python
import datetime
import pandas as pd
from py2neo import Graph
```

```python
pd.options.display.max_rows = None  # display all rows
pd.options.display.max_columns = None  # display all columsns
```

```python
graph = Graph("bolt://localhost:7687", user="neo4j", password="password")
```

```python
query = """
MATCH (n) RETURN n
"""
graph.run(query).to_data_frame()
```

```python
query = """
MATCH (l:Location)
RETURN l.location
"""
graph.run(query).to_data_frame()
```

```python
query = """
match (r:Restaurant)
return r.name, r.contactNumber, r.address
"""
graph.run(query).to_data_frame()
```

```python
query = """
match (r:Restaurant)<-[:at]-(l:Location)
return r.name, l.location
"""
graph.run(query).to_data_frame()
```

```python
query = """
match (r:Restaurant)-[:serves]->(:Buffet)
return r.name
"""
graph.run(query).to_data_frame()
```

```python
query = """
match (r:Restaurant)-[:serves]->(b:Buffet)
where b.price < 35
return r.name, b.price
"""
graph.run(query).to_data_frame()
```
